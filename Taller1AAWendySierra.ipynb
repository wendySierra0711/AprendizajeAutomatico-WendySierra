Realice el entrenamiento de un algoritmo de regresión logística utilizando sklearn


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

x = np.asarray(df[['protocol_type', 'service', 'flag', 'urgent', 'hot', 'count', 'serror_rate']])
x[0:7]

y = valuesClass
y [0:7]
# Codificación de variables categóricas a numéricas
from sklearn import preprocessing
colums = [0,1,2,3,4,5,6]
for x1 in colums:
    transform_tt = preprocessing.LabelEncoder()
    x[:,x1] = transform_tt.fit_transform(x[:,x1])

from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)


modelo_logistico = LogisticRegression()

# Entrenar el modelo utilizando los datos de entrenamiento
modelo_logistico.fit(x_train,y_train)
```

Realice la predicción del modelo generado utilizando el conjunto de datos de validación

```python
y_pred = modelo_logistico.predict(x_test)

# Calcular la precisión en el conjunto de prueba
accuracy = accuracy_score(y_test, y_pred)
print("Exactitud en el conjunto de validación:", accuracy)
```

Grafique la matriz de confusión

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
# matriz de confusión
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Clase Negativa', 'Clase Positiva'],
            yticklabels=['Clase Negativa', 'Clase Positiva'])
plt.xlabel('Predicción')
plt.ylabel('Etiqueta Verdadera')
plt.title('Matriz de Confusión')
plt.show()
```

Calcule la precisión, recall, f1 score y explique los resultados obtenidos

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Precisión:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```

Grafique la curva ROC y PR del modelo


```python
from sklearn.metrics import roc_curve, precision_recall_curve, auc
import matplotlib.pyplot as plt

# Supongamos que 'y_val' contiene las etiquetas verdaderas y 'y_scores' contiene las puntuaciones de decisión

# Calcular la curva ROC
fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Calcular la curva PR
precision, recall, _ = precision_recall_curve(y_test, y_pred)
pr_auc = auc(recall, precision)

# Crear una figura para las curvas ROC y PR
plt.figure(figsize=(12, 5))

# Graficar la curva ROC
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color='darkorange', lw=2, label='Curva ROC (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc="lower right")

# Graficar la curva PR
plt.subplot(1, 2, 2)
plt.plot(recall, precision, color='darkorange', lw=2, label='Curva PR (AUC = %0.2f)' % pr_auc)
plt.xlabel('Recall')
plt.ylabel('Precisión')
plt.title('Curva PR')
plt.legend(loc="lower left")

plt.tight_layout()
plt.show()
```
